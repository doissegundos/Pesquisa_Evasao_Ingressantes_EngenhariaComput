#### Perguntas relacionadas ao nivel de concordancia, são divididas em 3 classes: "Discordo", "Indiferente", "Concordo"

```
def concordancia(pergunta):
  plt.figure(figsize=(10, 6))
  sns.set_theme(style="whitegrid")
  data = df[pergunta].to_frame().reset_index()
  data.columns = ["index", "Resposta"]

  #Redefinindo os valores das classes
  data["Resposta"] = data["Resposta"].astype('str')
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("1", "Discordo"))
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("2", "Discordo"))
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("3", "Indiferente"))
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("4", "Concordo"))
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("5", "Concordo"))

  #Plotando 
  sns.countplot('Resposta',data=data,palette="mako", order = ["Discordo", "Indiferente","Concordo"])
  plt.title(pergunta) 
  plt.show()
```

#### Perguntas Relacionadas ao Tipo de Relevancia, podem ser divididas em 3 classes, sendo elas: "Irrelevante", "Indiferente, "Relevante"

```
def relevancia(pergunta):
  plt.figure(figsize=(10, 6))
  sns.set_theme(style="whitegrid")
  data = df[pergunta].to_frame().reset_index()
  data.columns = ["index", "Resposta"]
  #Redefinindo os valores das classes
  data["Resposta"] = data["Resposta"].astype('str')
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("Muito irrelevante", "Irrelevante"))
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("Irrelevante", "Irrelevante"))
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("Indiferente", "Indiferente"))
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("Relevante", "Relevante"))
  data["Resposta"] = data["Resposta"].apply(lambda x: x.replace("Muito relevante", "Relevante"))

  #Plotando 
  sns.countplot('Resposta',data=data,palette="mako", order = ["Irrelevante", "Indiferente","Relevante"])
  plt.title(pergunta) 
  plt.show()
```

```
relevancia('Qual a relevância dos aspectos que seguem para sua percepção de dificuldade nas disciplinas destacadas anteriormente: [Dificuldade na compreensão do conteúdo]')
relevancia('Qual a relevância dos aspectos que seguem para sua percepção de dificuldade nas disciplinas destacadas anteriormente: [Abordagem metodológica do professor]')
relevancia('Qual a relevância dos aspectos que seguem para sua percepção de dificuldade nas disciplinas destacadas anteriormente: [Avaliações excessivas]')
relevancia('Qual a relevância dos aspectos que seguem para sua percepção de dificuldade nas disciplinas destacadas anteriormente: [Pouca relação do conteúdo com a prática]')
```
